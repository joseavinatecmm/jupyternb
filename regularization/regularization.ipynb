# Regularización en Regresión Lineal

## Introducción
La regularización es una técnica utilizada en machine learning para prevenir el sobreajuste (overfitting) en modelos de regresión. En este notebook, exploraremos las técnicas de regularización **Ridge (L2)** y **Lasso (L1)**, y cómo se aplican en un modelo de regresión lineal.

## Teoría Matemática

### Regresión Lineal sin Regularización
En un modelo de regresión lineal, buscamos ajustar una función lineal a los datos:
$
\hat{y} = w^T x + b
$
donde:
- $w$ es el vector de pesos o coeficientes de regresión ($\beta$).
- $b$ es el sesgo.
- $x$ es el vector de características.
- $\hat{y}$ es la predicción.

La función de pérdida (MSE) es:
$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^n (y_i - \hat{y}_i)^2
$$

### Regularización Ridge (L2)
La regularización Ridge añade una penalización L2 a la función de pérdida:
$$
\text{Pérdida} = \text{MSE} + \lambda \sum_{j=1}^p w_j^2
$$
donde \(\lambda\) es el parámetro de regularización.

### Regularización Lasso (L1)
La regularización Lasso añade una penalización L1 a la función de pérdida:
$$
\text{Pérdida} = \text{MSE} + \lambda \sum_{j=1}^p |w_j|
$$
Lasso puede llevar algunos pesos a cero, realizando selección de características.

## Implementación en Python

### Paso 1: Generar un Dataset con Overfitting
Vamos a generar un dataset sintético para una regresión lineal simple, pero introduciremos ruido para simular overfitting.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

# Generar datos sintéticos
np.random.seed(42)
X = np.linspace(0, 10, 20)
y = 2 * X + 3 + np.random.normal(0, 2, size=X.shape)  # Recta con ruido

# Añadir overfitting con un polinomio de alto grado
poly = PolynomialFeatures(degree=15)
X_poly = poly.fit_transform(X.reshape(-1, 1))

# Ajustar un modelo lineal con overfitting
model_overfit = LinearRegression()
model_overfit.fit(X_poly, y)
y_overfit = model_overfit.predict(X_poly)

# Graficar los datos originales y la recta con overfitting
plt.scatter(X, y, color='blue', label='Datos originales')
plt.plot(X, y_overfit, color='red', label='Recta con overfitting')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Regresión Lineal con Overfitting')
plt.legend()
plt.show()
```

### Paso 2: Aplicar Regularización Ridge (L2)
Ahora aplicaremos regularización Ridge para evitar el overfitting.

```python
# Crear un modelo Ridge
model_ridge = Ridge(alpha=1.0)  # alpha es el parámetro de regularización lambda
model_ridge.fit(X_poly, y)
y_ridge = model_ridge.predict(X_poly)

# Graficar los datos originales, la recta con overfitting y la recta regularizada
plt.scatter(X, y, color='blue', label='Datos originales')
plt.plot(X, y_overfit, color='red', label='Recta con overfitting')
plt.plot(X, y_ridge, color='green', label='Recta regularizada (Ridge)')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Regularización Ridge (L2)')
plt.legend()
plt.show()
```

### Paso 3: Aplicar Regularización Lasso (L1)
Finalmente, aplicaremos regularización Lasso para comparar su efecto.

```python
# Crear un modelo Lasso
model_lasso = Lasso(alpha=0.1)  # alpha es el parámetro de regularización lambda
model_lasso.fit(X_poly, y)
y_lasso = model_lasso.predict(X_poly)

# Graficar los datos originales, la recta con overfitting y la recta regularizada
plt.scatter(X, y, color='blue', label='Datos originales')
plt.plot(X, y_overfit, color='red', label='Recta con overfitting')
plt.plot(X, y_lasso, color='purple', label='Recta regularizada (Lasso)')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Regularización Lasso (L1)')
plt.legend()
plt.show()
```

## Conclusión
- La regularización Ridge (L2) reduce la magnitud de los pesos, evitando valores extremos.
- La regularización Lasso (L1) lleva algunos pesos a cero, realizando selección de características.
- Ambas técnicas ayudan a prevenir el overfitting y mejorar la generalización del modelo.
